# Customers Segmentation with K-Means


```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Loading the file
df = pd.read_excel("C:/Users/cedri/Documents/BankChurners.xlsx", sheet_name="BankChurners")

# Slect the segmentation variables
features = ['Total_Trans_Amt', 'Total_Trans_Ct', 'Credit_Limit', 'Avg_Utilization_Ratio']
df_selected = df[features].dropna()

# Data normalization
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df_selected)

# K-Means application (3 segments)
kmeans = KMeans(n_clusters=3, random_state=42)
df_selected['Segment'] = kmeans.fit_predict(scaled_features)

# Visualization in 2D
plt.figure(figsize=(8, 6))
plt.scatter(df_selected['Total_Trans_Amt'], df_selected['Total_Trans_Ct'], c=df_selected['Segment'], cmap='Set2')
plt.xlabel("total of transactions Amount")
plt.ylabel("total of transactions number")
plt.title("Customers segmentation (K-Means)")
plt.grid(True)
plt.show()
```

# Save the segmentation in the new excel file

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Loading Data
df = pd.read_excel("C:/Users/cedri/Documents/BankChurners.xlsx", sheet_name="BankChurners")

# Slect the essential column
features = ['Total_Trans_Amt', 'Total_Trans_Ct', 'Credit_Limit', 'Avg_Utilization_Ratio']
df_segmentation = df[features].dropna()

# Normalization
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_segmentation)

# K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
segments = kmeans.fit_predict(X_scaled)

# Adding the segments to the original DataFrame
df['Segment'] = -1  # Value by default
df.loc[df_segmentation.index, 'Segment'] = segments

# Save in the new Excel files
df.to_excel("customers_segments.xlsx", index=False)

print(" Fichier 'customers_segments.xlsx' created with succes.")
```
